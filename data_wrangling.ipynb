### Answers - set 1

```python
# 1. Change all occurrences of the 'PG' rating to 'PG-13' across the dataset.
df_film['rating'].replace('PG', 'PG-13', inplace=True)

# 2. Fill in the missing values in the 'replacement_cost' column with the median value from that column.
df_film['replacement_cost'].fillna(df_film['replacement_cost'].median(), inplace=True)

# 3. Update the title 'ACADEMY DINOSAUR' to 'ACADEMY T-REX' wherever it appears in the dataset.
df_film['title'].replace('ACADEMY DINOSAUR', 'ACADEMY T-REX', inplace=True)

# 4. For each release year, calculate the average rental rate of films.
df_film.groupby('release_year')['rental_rate'].mean()

# 5. Count how many films fall under each rating category.
df_film['rating'].value_counts()

# 6. Replace any mention of 'Trailers' in the 'special_features' with 'Behind the Scenes'.
df_film['special_features'] = df_film['special_features'].str.replace('Trailers', 'Behind the Scenes')

# 7. Find the first film released in each year.
df_film.groupby('release_year').first()

# 8. For each rating, identify the film that has the highest replacement cost.
df_film.loc[df_film.groupby('rating')['replacement_cost'].idxmax()]

# 9. Replace zero values in the 'film_length' column with the average length of all films.
df_film['film_length'].replace(0, df_film['film_length'].mean(), inplace=True)

# 10. Calculate the total replacement cost of films for each rating.
df_film.groupby('rating')['replacement_cost'].sum()

# 11. Wherever the 'original_language_id' is missing, replace it with the value 1.
df_film['original_language_id'].fillna(1, inplace=True)

# 12. Identify the longest film released in each year.
df_film.loc[df_film.groupby('release_year')['film_length'].idxmax()]

# 13. Update the rental duration from 3 days to 5 days wherever applicable.
df_film['rental_duration'].replace(3, 5, inplace=True)

# 14. Find the shortest film for each language category.
df_film.loc[df_film.groupby('language_id')['film_length'].idxmin()]

# 15. Adjust the rental rate to 4.99 for all films released in 2006.
df_film.loc[df_film['release_year'] == 2006, 'rental_rate'] = 4.99

# 16. Determine the average length of films for each release year.
df_film.groupby('release_year')['film_length'].mean()

# 17. Change the 'special_features' to 'Behind the Scenes' for films with a replacement cost above 20.
df_film.loc[df_film['replacement_cost'] > 20, 'special_features'] = 'Behind the Scenes'

# 18. Find the last film in each rating category.
df_film.groupby('rating').last()

# 19. Fill in missing values in the 'rental_duration' column with the most common duration.
df_film['rental_duration'].fillna(df_film['rental_duration'].mode()[0], inplace=True)

# 20. Identify the film with the lowest rental rate for each release year.
df_film.loc[df_film.groupby('release_year')['rental_rate'].idxmin()]
```

```python

# 1. Adding a column
# Add a new column to df_rental that contains the number of days between rental_date and return_date.
df_rental['rental_duration'] = (df_rental['return_date'] - df_rental['rental_date']).dt.days

# 2. Concatenating two DataFrames on rows
# Concatenate df_rental and df_payment on rows (assuming they have the same columns or you want to stack them vertically).
df_concat_rows = pd.concat([df_rental, df_payment], ignore_index=True)

# 3. Concatenating two DataFrames on columns
# Concatenate df_rental and df_payment on columns (assuming they have the same number of rows).
df_concat_cols = pd.concat([df_rental, df_payment], axis=1)

# 4. Renaming columns
# Rename 'rental_date' to 'rental_start_date' and 'return_date' to 'rental_end_date' in df_rental.
df_rental.rename(columns={'rental_date': 'rental_start_date', 'return_date': 'rental_end_date'}, inplace=True)

# 5. Adding a row
# Add a new row to df_rental with the given values.
new_row = {'rental_id': 999, 'rental_start_date': '2024-09-01', 'rental_end_date': '2024-09-10', 'inventory_id': 10, 'customer_id': 1, 'staff_id': 1}
df_rental = df_rental.append(new_row, ignore_index=True)

# 6. Using lambda function
# Create a new column 'rental_period' in df_rental that contains the rental duration in months (using lambda).
df_rental['rental_period'] = df_rental.apply(lambda x: (x['rental_end_date'] - x['rental_start_date']).days // 30, axis=1)

# 7. Using apply
# Create a new column 'amount_due' in df_payment that is twice the value of 'amount'.
df_payment['amount_due'] = df_payment['amount'].apply(lambda x: x * 2)

# 8. Iterating on rows
# Print each row of df_rental along with its index.
for index, row in df_rental.iterrows():
    print(index, row)

# 9. Iterating on columns - using apply
# Apply a function to each column of df_payment to compute the number of non-null values in each column.
non_null_counts = df_payment.apply(lambda col: col.notnull().sum())

# 10. Use map
# Use map to create a new column in df_payment that labels payments greater than 50 as 'High' and others as 'Low'.
df_payment['amount_label'] = df_payment['amount'].map(lambda x: 'High' if x > 50 else 'Low')

# 11. For a str column - do some string operations
# In df_rental, create a new column 'rental_info' that combines 'inventory_id' and 'customer_id' as a string.
df_rental['rental_info'] = df_rental['inventory_id'].astype(str) + '-' + df_rental['customer_id'].astype(str)

# 12. .astype(str), some str column operations
# Convert 'rental_start_date' to string and then extract the year part.
df_rental['rental_year'] = df_rental['rental_start_date'].astype(str).str[:4]

# 13. Some date operations - .dt.date type question, pd.to_datetime, .dt.period etc
# Convert 'rental_start_date' to datetime and create a new column 'rental_month' with the month of the rental.
df_rental['rental_start_date'] = pd.to_datetime(df_rental['rental_start_date'])
df_rental['rental_month'] = df_rental['rental_start_date'].dt.to_period('M')

# 14. Fill missing values
# Fill missing values in the 'return_date' column with the value '2024-01-01'.
df_rental['return_date'].fillna('2024-01-01', inplace=True)

# 15. ffill, contrasting examples of forward fill, backward fill, mean and others
# Forward fill missing values in 'return_date' and backward fill missing values in 'rental_date'.
df_rental['return_date'] = df_rental['return_date'].ffill()
df_rental['rental_date'] = df_rental['rental_date'].bfill()

# 16. Taking cumulative - operation such that next row value is related to one previous row value
# Calculate the cumulative sum of 'amount' in df_payment.
df_payment['cumulative_amount'] = df_payment['amount'].cumsum()
```
